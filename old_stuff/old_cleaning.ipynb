```python
elections = pd.concat([pd.read_csv("1976-2018-house2.csv", encoding= 'unicode_escape'), pd.read_csv("1976-2018-senate.csv", encoding= 'unicode_escape')])
elections.sort_values(by=['year', 'state', 'office'], inplace=True)
elections = elections[elections.year >= 2012]
elections.dropna(subset=['candidate'], inplace=True)
elections[(elections.year == 2018) & (elections.state=='Wyoming')]
```

```python
earlier_data = pd.read_json("../bp_accounts/2012_14.json")
later_data = pd.read_json("../bp_accounts/2016_18.json")
df = pd.concat([earlier_data,later_data])
df['Body'].replace('U.S. House', "US House", inplace=True)
df['Body'].replace('U.S. Senate', "US Senate", inplace=True)
df.sort_values(by=['Year', 'State', 'Body'], inplace=True)
df.Name = df.Name.apply(lambda name: re.sub("\(.*\)", "", name).strip())
df = df[['State', 'Body', 'Party', 'Year', 'Name', 'Campaign Twitter',
                   'Personal Twitter', 'Official Twitter']]
df.dropna(how='all', inplace=True, subset=['Campaign Twitter',
                                                'Personal Twitter', 'Official Twitter'])
df.Party = df.Party.str.lower()
df[df.Name.str.contains('Zoe')]
# merged_df = df.merge(elections, how='outer',left_on=['Name'], right_on=['candidate'], indicator=True)
```
